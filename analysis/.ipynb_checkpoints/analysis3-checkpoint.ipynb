```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
day = pd.read_csv("../data/raw/day.csv")
day
```

```python
day_clean=day.drop(['instant','yr','weathersit',],axis=1)
day_clean.head()
```

```python
# Data analysis and visualization for RQ1
total_registered = day_clean['registered'].sum()
total_casual = day_clean['casual'].sum()

# Create a DataFrame for the data
df = pd.DataFrame({'type': ['registered', 'casual'], 'count': [total_registered, total_casual]})

# Create the bar plot
sns.barplot(x='type', y='count', data=df)
```

```python
# Data analysis and visualization for RQ1
daily_counts = day_clean.groupby('dteday')[['registered', 'casual']].sum()

# Reset the index to make the date a column
daily_counts = daily_counts.reset_index()

# Melt the data to create a "long" format with separate columns for the user types
daily_counts_long = pd.melt(daily_counts, id_vars=['dteday'], value_vars=['registered', 'casual'], var_name='type', value_name='count')

# Create the line plot
sns.lineplot(x='dteday', y='count', hue='type', data=daily_counts_long)
```

```python
# Data analysis and visualization for RQ2

# I creat a scatterplot to show that if there is any relation between temperature and user's numbers.
sns.scatterplot(x='temp', y='cnt', data=day_clean)
```

```python
# Data analysis and visualization for RQ3


# Create bar plot of average number of users on holidays vs. non-holidays
sns.barplot(x='holiday', y='cnt', data=day_clean)

# Create bar plot of average number of users on weekends vs. weekdays
sns.barplot(x='weekday', y='cnt', data=day_clean)
```

```python
# Data analysis and visualization for RQ3

# Create bar plot of average number of users on holidays vs. non-holidays
sns.barplot(x='holiday', y='cnt', data=day_clean)
```

```python
# Data analysis and visualization for RQ3

# Create box plot of number of users by holiday
sns.boxplot(x='holiday', y='cnt', data=day_clean)
```

### EDA - Summary and conclusions
1. From the total number of both types of users, registered users are much larger than casual users; from the data of each day, registered users are always more than casual users

2. The temperature is positively associated with the total number of users, the higher the temperature, the greater the proportion of more users, but after the temperature reaches 0.6, it starts to become an inverse association.

3. There is no correlation between the number of users and the day of the week.

4. There is a connection between the number of users and whether it is a holiday or not, and the number of users is more when it is not a holiday
